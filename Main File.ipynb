# US DATA BREACH ANALYSIS 2009-21
### This file consists of analysis of data breaches in the healthcare companies of the US from the year 2009-2021 


### In the first step we shall import our csv file of the data breaches

```python
import pandas as pd
df = pd.read_csv('data_breaches_all_clean.csv')
```

###  Printing the DataSet

```python
df
```

### In this step we have found the different healthcare sectors which were affected by DataBreaches

```python
df['Covered Entity Type'].unique()
```

### Here we filtered the data of companies which fall under 'Healthcare provider' entity

```python
entity = ['Healthcare Provider']
Healthcare=df[df['Covered Entity Type'].isin(entity)]
Healthcare
```

### Further we found overall people affected in the healthcare provider mentioned entity

```python
Affected_healthcare=Healthcare['Individuals Affected'].sum()
Affected_healthcare
```

### In further few steps we will be repeating the previous steps for different entity type

```python
entity = ['Business Associate']
Business=df[df['Covered Entity Type'].isin(entity)]
Business
```

```python
Affected_business=Business['Individuals Affected'].sum()
Affected_business
```

```python
entity = ['Health Plan']
HealthPlan=df[df['Covered Entity Type'].isin(entity)]
HealthPlan
```

```python
Affected_healthplan=HealthPlan['Individuals Affected'].sum()
Affected_healthplan
```

```python
entity = ['Healthcare Clearing House']
HealthCareCleaning=df[df['Covered Entity Type'].isin(entity)]
HealthCareCleaning
```

```python
Affected_healthcarecleaning=HealthCareCleaning['Individuals Affected'].sum()
Affected_healthcarecleaning
```

### As we calculated the number of people affected in each entity type, we created a new dataframe with columns of 'Entity Type' and 'Number of Affected' 

```python
Analysis={'Entity Type':['Healthcare','Business','HealthPlan','HealthCareCleaning'],'Number of Affected':[Affected_healthcare,Affected_business,Affected_healthplan,Affected_healthcarecleaning]}
```

```python
df_1 = pd.DataFrame(Analysis,index=['Healthcare','Business','HealthPlan','HealthCareCleaning'])
df_1
```

### Now, we Sketched a Pie diagram which represents our above created dataframe 

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
df_1.plot.pie(y='Number of Affected',figsize=(5,5),autopct='%1.1f%%')
plt.legend(loc='lower left',shadow=True)

```

#### We can see that the most people affected belongs to 'Healthcare' and 'HealthPlan' Entity type. As we know robbers tend to target banks, rather than newspaper stands, cybercriminals are on the lookout for ‘big ticket’ targets, which can offer the most value.large medical organizations hold the most amount of data, they also have larger cybersecurity budgets, and are generally part of hospital networks, meaning they share cybersecurity best practices, and closed intranets with other hospitals in the area.Smaller enterprises may have less user data, but they also have lower security budgets and less complex and up-to-date cybersecurity solutions.
